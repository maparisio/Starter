### Part 1: Simple DataFrame clean up
1. Create a Data Frame with the CSV file
2. Separated Data Frames: US and Rest of the World
3. Separated the Location by comma to obtain:
   USA: Complex ID, Launch Site, State, Country
   Rest of World: Complex ID, Launch Site, Country

```python
import pandas as pd

#Specify the file path and encoding
file_path = 'Resources\space_missions.csv'
encoding = 'ISO-8859-1'  # or 'latin1'

#Load the CSV file with the specified encoding
spaceMissions_df = pd.read_csv(file_path, encoding=encoding)

#Display the FULL loaded DataFrame
spaceMissions_df
```

```python
#Get unique values in the "Launch Site" column.  Will need to fix rows that are missing a country, except for "Pacific Ocean"
unique_launch_sites = spaceMissions_df['Location'].unique()
unique_launch_sites
```

```python
#Append Countries to locations that do not have them

#Replace 'Barents Sea' with 'Russia' in the 'Location' column
spaceMissions_df['Location'] = spaceMissions_df['Location'].str.replace('Barents Sea', 'Russia', case=False)

# Locate and update rows with 'Kauai'
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Kauai'), 'Location'] += ', USA'

#Replace 'Pacific Missile Range Facility' with 'Hawaii' in the 'Location' column
spaceMissions_df['Location'] = spaceMissions_df['Location'].str.replace('Pacific Missile Range Facility', 'Hawaii', case=False)

# Locate and update rows with incomplete New Mexico
#spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Vertical Launch Area'), 'Location'] += ', USA'

# Replace 'Kodiak' with 'Alaska' in the 'Location' column
spaceMissions_df['Location'] = spaceMissions_df['Location'].str.replace('Kodiak', 'Alaska', case=False)

#Replace 'French Guiana' with 'France' in the 'Location' column
spaceMissions_df['Location'] = spaceMissions_df['Location'].str.replace('French Guiana', 'France', case=False)

#Replace 'MaranhÃ£o' with 'Brazil' in the 'Location' column
spaceMissions_df['Location'] = spaceMissions_df['Location'].str.replace('MaranhÃ£o', 'Brazil', case=False)

# Locate and update rows with 'Yellow Sea'
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Yellow Sea'), 'Location'] += ', China'

# Locate and update rows with 'Yellow Sea'
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Uchinoura Space Center'), 'Location'] += ', Japan'

#Add space to get Russia to Svobodny Cosmodrome:
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Svobodny Cosmodrome, Russia'), 'Location'] = 'Svobodny Cosmodrome, , Russia'
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Taiyuan Satellite Launch Center, China'), 'Location'] = 'Taiyuan Satellite Launch Center, , China'
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Jiuquan Satellite Launch Center, China'), 'Location'] = 'Jiuquan Satellite Launch Center, , China'
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Xichang Satellite Launch Center, China'), 'Location'] = 'Xichang Satellite Launch Center, , China'
# Locate and update rows with 'Shahrud Missile Test Site' and does not contain "Iran" as country
spaceMissions_df.loc[spaceMissions_df['Location'].str.contains('Shahrud Missile Test Site') & ~spaceMissions_df['Location'].str.contains('Iran'), 'Location'] += ', Iran'

# Replace 'Algeria' with 'France' in the 'Location' column
spaceMissions_df['Location'] = spaceMissions_df['Location'].str.replace('Algeria', 'France', case=False)

spaceMissions_df.loc[
    spaceMissions_df['Location'].str.contains('Vertical Launch Area') & 
    ~spaceMissions_df['Location'].str.contains('USA'), 
    'Location'
] += ', USA'
```

```python
#List of locations after appending missing Countries
appended_locations = spaceMissions_df['Location'].unique()
appended_locations
```

###
# Data Frame with US Space Missions
###

### US Space Missions DataFrame with Location intact

```python
#Separate launches into two DataFrames
#This will be the US Space Missions and later on we will do the rest of the world's Space Missions
us_launches = spaceMissions_df[spaceMissions_df['Location'].str.contains('USA')]

#Display the US DataFrame
us_launches
```

### Cleaned US Space Missions - Split Location Column

```python
#Split the "Location" column into separate parts
location_parts = us_launches['Location'].str.split(', ', expand=True)

#Rename the new columns dynamically based on the number of columns
new_columns = ['Complex ID', 'Launch Site', 'Territory', 'Country', 'Column_5']
location_parts.columns = new_columns

#Concatenate the new columns with the original DataFrame
cleaned_US_df = pd.concat([us_launches, location_parts], axis=1)

#Drop the original "Location" column and the unnecessary column
cleaned_US_df = cleaned_US_df.drop(['Location', 'Column_5'], axis=1, errors='ignore')

#Display the updated DataFrame
cleaned_US_df
```

```python
#Count of column data
cleaned_US_df.count()
```

```python
#What territory had a Space Mission Launch
unique_territory = cleaned_US_df['Territory'].unique()
territory_counts = cleaned_US_df['Territory'].nunique()

#print the info
print(f"The USA had space missions in a total of {territory_counts} states and/or territories: {unique_territory}")
```

```python
#Group by state and count. Set 'State' as index
missions_per_territory = cleaned_US_df.groupby('Territory').size().reset_index(name='Mission Count').set_index('Territory')

#Display DataFrame to show the missions per Country
missions_per_territory
```

```python
#Filter missions based on success and failure for the US space missions
successful_missions = cleaned_US_df[cleaned_US_df['MissionStatus'] == 'Success']
failed_missions = cleaned_US_df[cleaned_US_df['MissionStatus'] == 'Failure']

#Display the results
successful_missions
```

```python
print(f"The total of successful Space Missions in the US were: {successful_missions.shape[0]}")
```

### Failed Space Missions in the US

```python
#Print Failed Missions DataFrame
failed_missions
```

```python
print(f"The total of failed Space Missions in the US were: {failed_missions.shape[0]}")
```

```python
# Group by state and count missions per state and MissionStatus
missions_per_territory = cleaned_US_df.pivot_table(index='Territory', columns='MissionStatus',aggfunc='size', fill_value=0)

#Print Data Frame
missions_per_territory
```

# DataFrame with Space Missions from the Rest of the World
###

```python
#Show Space Missions of the world excluding the US
rest_of_world_launches = spaceMissions_df[~(spaceMissions_df['Location'].str.contains('USA') )]
rest_of_world_launches
```

```python
world_uniquelocations = rest_of_world_launches['Location'].unique()
world_uniquelocations
```

### Cleaned Rest of World Space Missions - Split Location Column 

```python
# Split the "Location" column into separate parts
new_worldcolumns = rest_of_world_launches['Location'].str.split(', ', expand=True)

# Rename the new columns dynamically based on the number of columns
new_world_columns = ['Complex ID', 'Launch Site', 'Country', 'Column_4']
new_worldcolumns.columns = new_world_columns

# Concatenate the new columns with the original DataFrame for the rest of the world
cleaned_world_df = pd.concat([rest_of_world_launches, new_worldcolumns], axis=1)

# Drop the original "Location" column and the unnecessary columns
cleaned_world_df = cleaned_world_df.drop(['Location', 'Column_4'], axis=1, errors='ignore')

# Display the updated DataFrame
cleaned_world_df
```

```python
#Count of column data
cleaned_world_df.count()
```

```python
#What Countries had a Space Mission Launch
unique_countries = cleaned_world_df['Country'].unique()
country_counts = cleaned_world_df['Country'].nunique()

#print the info
print(f"A total of {country_counts} Countries had Space Missions in the rest of the world: {unique_countries}")
```

```python
# Group by country and count missions per country, set "Country" as index
missions_per_country = cleaned_world_df.groupby('Country').size().reset_index(name='Mission Count').set_index('Country')

# Display DataFrame to show the missions per Country
missions_per_country
```

```python
#Filter missions based on success and failure for the 'Rest of the World' space missions
successful_w_missions = cleaned_world_df[cleaned_world_df['MissionStatus'] == 'Success']
failed_w_missions = cleaned_world_df[cleaned_world_df['MissionStatus'] == 'Failure']

#Display successful results
successful_w_missions
```

```python
#Display failed results
failed_w_missions
```

```python
#Group by state and count missions per state and MissionStatus
missions_per_countryNUS = cleaned_world_df.pivot_table(index='Country', columns='MissionStatus',aggfunc='size', fill_value=0)

#Print Data Frame
missions_per_countryNUS
```
